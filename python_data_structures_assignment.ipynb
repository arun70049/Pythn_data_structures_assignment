```python
# Q 1. Discuss string slicing and provide examples.

''' ANS: String slicing is a powerful feature in Python that allows us to
extract parts of a string by specifying a start, stop, and step value.
 It's a concise way to get substrings or manipulate strings.
 
Basic Syntax
The syntax for slicing a string is:
string[start:stop:step] 

start: The index where the slice starts (inclusive).
stop: The index where the slice ends (exclusive).
step: The step between each index to include in the slice.
If any of these parameters are omitted, Python provides default values:

start defaults to 0 if omitted.
stop defaults to the end of the string if omitted.
step defaults to 1 if omitted.

Examples
examples to illustrate how string slicing works.

Basic Slicing '''


s = "Hello, World!"
print(s[0:5])  # Output: Hello

# Here, s[0:5] extracts the substring from index 0 up to,
but not including,index 5.

# Omitting Parameters


s = "Hello, World!"
print(s[:5])   # Output: Hello
print(s[7:])   # Output: World!
print(s[:])    # Output: Hello, World!

''' s[:5] gets the substring from the beginning up to index 5.
s[7:] gets the substring from index 7 to the end.
s[:] gets the whole string. '''

# Using Step

s = "Hello, World!"
print(s[::2])  # Output: Hlo ol!
print(s[1:10:2])  # Output: el,W

# s[::2] takes every second character from the string.
# s[1:10:2] takes every second character from index 1 to 9.

# Negative Indices

s = "Hello, World!"
print(s[-6:-1])  # Output: World
print(s[::-1])   # Output: !dlroW ,olleH

""" s[-6:-1] extracts the substring from the 6th character from the end up to,
but not including, the last character.
 s[::-1] reverses the entire string.
 Slicing with Different Steps """


s = "1234567890"
print(s[::3])  # Output: 1470
print(s[1::3])  # Output: 2589

# s[::3] extracts every 3rd character starting from index 0.
# s[1::3] extracts every 3rd character starting from index 1.

 # Insights:
# String slicing is a versatile tool in 
Python that lets you efficiently extract and manipulate parts of strings.
```

```python
# Q 2. Explain the key features of lists in  python.

''' ANS: Lists in Python are a versatile and widely used data structure
that can store a collection of items. They are particularly useful due to their 
dynamic nature and the broad range of operations they support.
Here are the key features of lists in Python:

1. Ordered
Lists maintain the order of elements.
This means that the order in which you add items to the list is the order in which 
they are stored and retrieved.

Example: '''

my_list = [10, 20, 30]
print(my_list[0])  # Output: 10

''' 2. Mutable
Lists are mutable, meaning that you can change their content after they have been created. You can add, remove, or modify items in a list.

Examples:  '''


# Modifying elements
my_list = [10, 20, 30]
my_list[1] = 25
print(my_list)  # Output: [10, 25, 30]

# Adding elements
my_list.append(40)
print(my_list)  # Output: [10, 25, 30, 40]

# Removing elements
my_list.remove(25)
print(my_list)  # Output: [10, 30, 40]

''' 3. Dynamic Size
Lists in Python can grow and shrink in size as needed.
We donâ€™t need to define their size when you create them.

Example: '''

my_list = []
my_list.append(1)
my_list.append(2)
print(my_list)  # Output: [1, 2]

''' 4. Heterogeneous
Lists can contain items of different data types, including other lists.

Example: '''

my_list = [1, "Hello", 3.14, [2, 4, 6]]
print(my_list[3])  # Output: [2, 4, 6]


''' 5. Indexing and Slicing
Lists support indexing and slicing, allowing you to access and modify subsets 
of the list. '''

# Examples: 

my_list = [10, 20, 30, 40, 50]
print(my_list[1:4])  # Output: [20, 30, 40]

# Negative indexing
print(my_list[-1])  # Output: 50
print(my_list[-3:-1])  # Output: [30, 40]

''' 6. Iterable
Lists are iterable, which means you can use them in loops and other contexts where
iteration is required. '''

# Example:


my_list = [1, 2, 3, 4, 5]
for item in my_list:
    print(item)
    
''' 7. Various Built-in Methods
Python lists come with a number of built-in methods for common operations: '''

# append(x): Adds an item to the end of the list.
#       extend(iterable): Extends the list by appending elements from an iterable.
# insert(i, x): Inserts an item at a specified position.
# remove(x): Removes the first occurrence of an item.
# pop([i]): Removes and returns an item at the specified position (or the last item if no index is provided).
# clear(): Removes all items from the list.
# index(x[, start[, end]]): Returns the index of the first occurrence of an item.
# count(x): Returns the number of occurrences of an item.
# sort(key=None, reverse=False): Sorts the items in place.
# reverse(): Reverses the items in place.
# copy(): Returns a shallow copy of the list.

#Examples:

my_list = [1, 2, 3, 4, 5]

my_list.append(6)
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]

my_list.insert(2, 10)
print(my_list)  # Output: [1, 2, 10, 3, 4, 5, 6]

my_list.remove(10)
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]

print(my_list.pop())  # Output: 6
print(my_list)  # Output: [1, 2, 3, 4, 5]

#  Insights:
    
''' Python lists are highly flexible and useful for managing collections of items.
Their key features include their ordered nature, mutability, dynamic size,
ability to hold heterogeneous data types, support for indexing and slicing,
and a rich set of built-in methods. '''
```

```python
# Q 3. Describe how to access, modify and delete elements in a list with examples.

''' ANS: In Python, you can access, modify, and delete elements in a list using 
various techniques.
Explanation of each operation with examples:

1. Accessing Elements
We can access elements in a list using indexing.
Python lists are zero-indexed, which means the first element has an index of 0. '''

# Examples:

my_list = ['apple', 'banana', 'cherry']

# Accessing elements using positive indices
print(my_list[0])  # Output: apple
print(my_list[1])  # Output: banana

# Accessing elements using negative indices (counts from the end)
print(my_list[-1])  # Output: cherry
print(my_list[-2])  # Output: banana

# Accessing a range of elements using slicing
print(my_list[1:3])  # Output: ['banana', 'cherry']

''' 2. Modifying Elements
We can modify elements in a list by assigning a new value to a specific index.
We can also use slicing to modify multiple elements at once. '''

# Examples:

my_list = [10, 20, 30, 40, 50]

# Modifying a single element
my_list[2] = 35
print(my_list)  # Output: [10, 20, 35, 40, 50]

# Modifying elements using slicing
my_list[1:4] = [25, 30, 45]
print(my_list)  # Output: [10, 25, 30, 45, 50]

# Modifying with a different length
my_list[1:3] = [27, 33, 37]
print(my_list)  # Output: [10, 27, 33, 37, 45, 50]

''' 3. Deleting Elements
We can delete elements from a list using several methods: '''

# Using del Statement: Deletes an element by index or a slice of elements.
# Using remove() Method: Removes the first occurrence of a specific value.
# Using pop() Method: Removes and returns an element at a specific index or
#                     the last element if no index is specified.
# Using clear() Method: Removes all elements from the list. 

#Examples:

my_list = ['apple', 'banana', 'cherry', 'date']

# Deleting an element by index
del my_list[1]
print(my_list)  # Output: ['apple', 'cherry', 'date']

# Deleting a slice of elements
del my_list[0:2]
print(my_list)  # Output: ['date']

# Removing an element by value
my_list = ['apple', 'banana', 'cherry']
my_list.remove('banana')
print(my_list)  # Output: ['apple', 'cherry']

# Removing and returning an element by index
popped_element = my_list.pop(0)
print(popped_element)  # Output: apple
print(my_list)  # Output: ['cherry']

# Removing and returning the last element
last_element = my_list.pop()
print(last_element)  # Output: cherry
print(my_list)  # Output: []

# Clearing all elements
my_list = ['apple', 'banana', 'cherry']
my_list.clear()
print(my_list)  # Output: []
# Insights
# Accessing Elements: Use indexing (list[index]) or slicing (list[start:stop]).
# Modifying Elements: Assign new values using indexing or slicing (list[index] = value or list[start:stop] = new_values).
# Deleting Elements: Use del list[index], list.remove(value), list.pop([index]), or list.clear().
```

```python
# Q 5. Describe the key features of sets and provide examples of their use.

''' ANS: Sets in Python are a powerful and versatile data structure used to store
unordered collections of unique items.
They are particularly useful when we need to manage collections of items without 
duplicates and perform operations like union, intersection, and difference.

Key features of sets, along with examples of their use:

Key Features of Sets
(a) Unordered Collection

Sets do not maintain any order of elements.
The items in a set do not have an index, and their order can change.
'''

# Example:

my_set = {1, 2, 3, 4}
print(my_set)  # Output: {1, 2, 3, 4}

''' (b)Unique Elements

Sets automatically ensure that all elements are unique.
Duplicate values are ignored when adding elements to a set. '''

#Example:

my_set = {1, 2, 3, 2, 1}
print(my_set)  # Output: {1, 2, 3}

''' (c) Mutable

Sets are mutable, meaning we can add or remove elements after creation.
However, the elements themselves must be immutable (e.g., numbers, strings, tuples).'''

#Example:

my_set = {1, 2, 3}
my_set.add(4)
my_set.remove(2)
print(my_set)  # Output: {1, 3, 4}

''' (d) No Duplicate Values

Sets automatically handle duplicates and only store unique values. '''

# Example:

my_set = {1, 1, 2, 3, 3}
print(my_set)  # Output: {1, 2, 3}

''' (e)Dynamic Size

Sets can grow and shrink in size dynamically as we add or remove elements. '''

# Example:

my_set = {1, 2}
my_set.add(3)
my_set.discard(1)
print(my_set)  # Output: {2, 3}

''' Set Operations

Sets support various set operations like union, intersection, difference,
and symmetric difference.'''

# Examples:

# Union: Combines elements from two sets, excluding duplicates.

set1 = {1, 2, 3}
set2 = {3, 4, 5}
union_set = set1 | set2
print(union_set)  # Output: {1, 2, 3, 4, 5}

# Intersection: Finds common elements between two sets.

set1 = {1, 2, 3}
set2 = {3, 4, 5}
intersection_set = set1 & set2
print(intersection_set)  # Output: {3}

# Difference: Finds elements that are in the first set but not in the second.


set1 = {1, 2, 3}
set2 = {3, 4, 5}
difference_set = set1 - set2
print(difference_set)  # Output: {1, 2}

# Symmetric Difference: Finds elements that are in either of the sets but not in both.

set1 = {1, 2, 3}
set2 = {3, 4, 5}
sym_diff_set = set1 ^ set2
print(sym_diff_set)  # Output: {1, 2, 4, 5}

''' Set Comprehension

Like list comprehensions, sets can be created using set comprehensions for more concise code.
'''
# Example:

squares = {x**2 for x in range(6)}
print(squares)  # Output: {0, 1, 4, 9, 16, 25}

# Examples of Set Use Cases
# Removing Duplicates from a List

numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = set(numbers)
print(unique_numbers)  # Output: {1, 2, 3, 4, 5}

# Membership Testing


fruits = {'apple', 'banana', 'cherry'}
print('apple' in fruits)  # Output: True
print('orange' in fruits)  # Output: False

# Finding Common Elements in Multiple Collections

students_math = {'Alice', 'Bob', 'Charlie'}
students_science = {'Bob', 'Charlie', 'David'}
common_students = students_math & students_science
print(common_students)  # Output: {'Charlie', 'Bob'}

# Set Operations for Data Analysis

set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
difference = set_a - set_b
print(difference)  # Output: {1, 2, 3}

# Insights
''' Sets in Python are highly useful for managing collections of unique items with
efficient membership testing and set operations.
Their key features include being unordered, containing unique elements, 
and supporting dynamic sizes and various set operations.'''
```

```python
# Q 7. Describe how to add, modify and delete items in a dictionary with examples.
''' ANS: Overview of how to add, modify,and delete items in a Python dictionary:

1. Adding Items
We can add new key-value pairs to a dictionary by simply assigning a valueto a new key.
'''

# Example:

my_dict = {'a': 1, 'b': 2}

# Adding a new key-value pair
my_dict['c'] = 3
print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 3}

''' 2. Modifying Items
To modify the value associated with an existing key,
assign a new value to that key.'''

# Example:

my_dict = {'a': 1, 'b': 2}

# Modifying the value of an existing key
my_dict['b'] = 20
print(my_dict)  # Output: {'a': 1, 'b': 20}

''' 3. Deleting Items
We can delete items using several methods:'''

# del Statement: Removes a key-value pair by key.
# pop() Method: Removes and returns the value associated with a key.
# popitem() Method: Removes and returns an arbitrary key-value pair (in Python 3.7+ it removes the last inserted pair).
# clear() Method: Removes all items from the dictionary.

# Examples:

my_dict = {'a': 1, 'b': 2, 'c': 3}

# Using del statement
del my_dict['a']
print(my_dict)  # Output: {'b': 2, 'c': 3}

# Using pop() method
value = my_dict.pop('b')
print(value)    # Output: 2
print(my_dict)  # Output: {'c': 3}

# Using popitem() method
item = my_dict.popitem()
print(item)     # Output: ('c', 3)
print(my_dict)  # Output: {}

# Using clear() method
my_dict = {'a': 1, 'b': 2}
my_dict.clear()
print(my_dict)  # Output: {}

# Insights
''' Add Items: dict[key] = value
Modify Items: dict[key] = new_value
Delete Items: del dict[key], dict.pop(key), dict.popitem(), dict.clear() '''
```
